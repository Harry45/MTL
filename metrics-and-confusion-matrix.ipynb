## Plot Confusion Matrix

```python
def plot_cm(ytrue: np.ndarray, ypred: np.ndarray, title: str, classes_types: list, normalize: bool, cmap: object, **kwargs) -> None:
    """Plot the confusion matrix given the targets and the predictions

    Args:
        ytrue (np.ndarray): The true targets from the data
        ypred (np.ndarray): The predicted targets from the model
        metrics (list): The metrics to be plotted
        classes_types (list): The different types of classes of data
        show_img (bool) : Option to display the image
        normalize (bool): The option to normalize the confusion matrix
        cmap (object): The color map to use for the plot
        output_fn (str): Path director to save the file
    """

    cf_matrix = confusion_matrix(y_true=ytrue, y_pred=ypred)

    if normalize:
        cf_matrix = cf_matrix.astype('float') / cf_matrix.sum(axis=1)[:, np.newaxis]

    # title = f'Confusion matrix: Accuracy = {metrics[0]:.2f}, Balanced Accuracy = {metrics[1]:.2f}'

    plt.figure(figsize=(9, 8))
    plt.imshow(cf_matrix, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=font_size, y=1.05)
    colorbar = plt.colorbar(fraction=0.046, pad=0.04)
    colorbar.ax.tick_params(labelsize=font_size)
    tick_marks = np.arange(len(classes_types))
    plt.xticks(tick_marks, classes_types, rotation=45)
    plt.yticks(tick_marks, classes_types)
    plt.tick_params(axis='x', labelsize=font_size)
    plt.tick_params(axis='y', labelsize=font_size)
    plt.ylabel('True label', fontsize=font_size)
    plt.xlabel('Predicted label', fontsize=font_size)
    plt.tight_layout()

    # add the values to the confusion matrix (and adjust color based on numbers)
    for i, j in itertools.product(range(cf_matrix.shape[0]), range(cf_matrix.shape[1])):
        if not normalize:

            color = "white" if (cf_matrix[i, j] / np.amax(cf_matrix) >= 0.70) else "black"

            plt.text(
                j, i, "{:d}".format(cf_matrix[i, j]),
                horizontalalignment="center", color=color, fontsize=font_size)

        else:
            color = "white" if (cf_matrix[i, j] >= 0.70) else "black"

            plt.text(j, i, "{:.2f}".format(cf_matrix[i, j]),
                     horizontalalignment="center", color=color, fontsize=font_size)

    if 'filename' in kwargs:
        # plt.savefig(kwargs['filename'] + '.pdf', bbox_inches='tight')
        
        if normalize:
            plt.savefig('plots/norm/' + kwargs['filename'] + '.png', bbox_inches='tight')
            plt.savefig('plots/norm/' + kwargs['filename'] + '.pdf', bbox_inches='tight')
        else:
            plt.savefig('plots/unnorm/' + kwargs['filename'] + '.png', bbox_inches='tight')
            plt.savefig('plots/unnorm/' + kwargs['filename'] + '.pdf', bbox_inches='tight')
        
    plt.close()
```

# Multi-Task Tree

```python
def find_labels(tasks: dict, labels: dict):
    """
    Example of input:
    
    tasks = {'task_1': [1, 0, 0], 'task_2': [1, 0, 0]...}
    """
    labels = {k:np.asarray(v) for k,v in labels.items()}
    tasks = {k:np.asarray(v) for k,v in tasks.items()}
    
    record_labels = []
    
    record_labels.append(labels['task_1'][tasks['task_1'] == 1])
    
    if tasks['task_1'][0] == 1:
        
        record_labels.append(labels['task_2'][tasks['task_2'] == 1])
        record_labels.append(labels['task_4'][tasks['task_4'] == 1])
        
    elif tasks['task_1'][1] == 1:
        record_labels.append(labels['task_3'][tasks['task_3'] == 1])
        
        if tasks['task_3'][0] == 1:
            record_labels.append(labels['task_5'][tasks['task_5'] == 1])
            record_labels.append(labels['task_4'][tasks['task_4'] == 1])
            
        else:
            record_labels.append(labels['task_6'][tasks['task_6'] == 1])
            record_labels.append(labels['task_7'][tasks['task_7'] == 1])
            
            if tasks['task_7'][0] == 1:
                record_labels.append(labels['task_8'][tasks['task_8'] == 1])
                record_labels.append(labels['task_9'][tasks['task_9'] == 1])
                record_labels.append(labels['task_10'][tasks['task_10'] == 1])
                record_labels.append(labels['task_4'][tasks['task_4'] == 1])
                
            else:
                record_labels.append(labels['task_10'][tasks['task_10'] == 1])
                record_labels.append(labels['task_4'][tasks['task_4'] == 1])

                       
    return np.asarray(record_labels).reshape(-1)
```

```python
labels = {

    'task_1' : ['Smooth', 'Featured or Disk', 'Artifact'],
    'task_2' : ['Round', 'In Between', 'Cigar Shaped'],
    'task_3' : ['Edge On Disk (Yes)', 'Edge On Disk (No)'],
    'task_4' : ['Merging (Merger)', 'Merging (Major Disturbance)', 'Merging (Minor Disturbance)', 'Merging (None)'],
    'task_5' : ['Bulge (Rounded)', 'Bulge (Boxy)', 'Bulge (None)'],
    'task_6' : ['No Bar', 'Weak Bar', 'Strong Bar'],
    'task_7' : ['Spiral Arms (Yes)', 'Spiral Arms (No)'],
    'task_8' : ['Spiral Winding (Tight)', 'Spiral Winding (Medium)', 'Spiral Winding (Loose)'],
    'task_9' : ['Spiral Arms (1)', 'Spiral Arms (2)', 'Spiral Arms (3)', 'Spiral Arms (4)', 'Spiral Arms (More Than 4)', 'Spiral Arms (cannot tell)'],
    'task_10': ['Central Bulge (None)', 'Central Bulge (Small)', 'Central Bulge (Moderate)', 'Central Bulge (Large)', 'Central Bulge (Dominant)']
         
}
```

```python
tasks = {

    'task_1' : [0, 1, 0],
    'task_2' : [0, 1, 0],
    'task_3' : [0, 1],
    'task_4' : [0, 1, 0, 0],
    'task_5' : [1, 0, 0],
    'task_6' : [0, 1, 0],
    'task_7' : [1, 0],
    'task_8' : [0, 1, 0],
    'task_9' : [0, 0, 1, 0, 0, 0],
    'task_10': [0, 1, 0, 0, 0]
         
}
```

```python
find_labels(tasks, labels)
```
